```python
"""
Name: question_4.py

Description:

Draw shapes: triangle, square, pentagon, hexagon, octagon.

Author:

Date:

Here is a list of colours: https://trinket.io/docs/colors, click on the 
colour to see the turtle name

"""

from turtle import *

# Function to draw a triangle
def draw_triangle(side_length):
    
    for counter in range(3):
        fd(side_length)
        lt(120)
        
penup()
setpos(-260, 0)
pendown()
pensize(2)
pencolor("Green")
draw_triangle(30)


# Function to draw a square

def draw_square(side_length):
    
    for counter in range(4):
        fd(side_length)
        lt(90)


# place the turtle on the left hand side of your canvas
penup()
setpos(-210, 0)
pendown()

# set the pen width
pensize(2)

# set the pen colour to a green colour
pencolor("Blue")

# Draw square
draw_square(50)


# Function to draw a pentagon
penup()
setpos(-120, 0)
pendown()
def draw_pentagon(side_length):
    
    for counter in range(5):
        fd(side_length)
        lt(72)

pensize(2)
pencolor("Green")
draw_pentagon(70)


penup()
setpos(20, 0)
pendown()
def draw_hexagon(side_length):
    
    for counter in range(6):
        fd(side_length)
        lt(60)

pensize(2)
pencolor("Red")
draw_hexagon(80)


penup()
setpos(200, 0)
pendown()
def draw_octagon(side_length):
    
    for counter in range(8):
        fd(side_length)
        lt(45)
pensize(2)
pencolor("Pink")
draw_octagon(81)
exitonclick()
```

```python
"""
Name: question_5.py

Description:

Draw shapes.

Author: Rae Harbird

Date: August 2018

Here is a list of colours: https://trinket.io/docs/colors, click on the 
colour to see the turtle name
"""

from turtle import *


# Function to draw a polygon

def draw_n_gon(sides, side_length): 
 

    for counter in range(sides):
        fd(side_length)
        lt(180-(sides-2)*180/sides)

    

# place the turtle on the left hand side of your canvas
penup()
setpos(-300,0)
pendown()

# set the pen width
pensize(2)

# set the pen colour to a green colour
pencolor("DarkOliveGreen4")

# Draw triangle
draw_n_gon(8, 50)

# leave the turtle on the screen until the user clicks in the screen
exitonclick()

```

```python
#Question 6


from turtle import *

def draw_n_gon(sides, side_length): 
 

    for counter in range(sides):
        fd(side_length)
        lt(180-(sides-2)*180/sides)

#rotate the polygon
def draw_polygon_rotate(side,side_length,repeats):
    
    for i in range(repeats):
        draw_n_gon(side,side_length)
        left(360/repeats)

        
penup()
setpos(-300,0)
pendown()
pensize(2)
pencolor("DarkOliveGreen4")

draw_polygon_rotate(3,50,10)

penup()
setpos(-150,0)
pendown()
pensize(2)
pencolor("Blue")
draw_polygon_rotate(4,70,10)


penup()
setpos(150,0)
pendown()
pensize(2)
pencolor("Green")
draw_polygon_rotate(5,90,10)

exitonclick()
```

```python
#Question 7 spirals
from turtle import *


def draw_spiral(sides,side_lenth,repeats):
    for counter in range (repeats*sides):
        fd(side_lenth)
        lt(180-(sides-2)*180/sides)      #直接写360/n 闭合的多边形
        
        side_lenth+=10


penup()
setpos(-300,0)
pendown()
pensize(2)
pencolor("DarkOliveGreen4")
   
draw_spiral(3,20,4)

penup()
setpos(-100,0)
pendown()
pensize(2)
pencolor("Blue")
draw_spiral(4,30,4)
exitonclick()  
```
